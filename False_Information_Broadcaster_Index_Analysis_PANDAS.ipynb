## VaccinItaly Data

```python
vaccinitaly_df = pd.read_csv("tweets_processed_1M.csv",
                             encoding="utf-8",
                             lineterminator='\n',
                             low_memory=False)
```

```python
ids_to_fix = ["in_reply_to_status_id", "retweeted_status_id", "quoted_status_id"]
vaccinitaly_df[ids_to_fix] = vaccinitaly_df[ids_to_fix].fillna(-1).astype(int)
```

```python
vaccinitaly_df
```

```python
urls_to_fix = ["urls", "quoted_urls"]
vaccinitaly_df[urls_to_fix] = vaccinitaly_df[urls_to_fix].fillna("missing;")
```

```python
vaccinitaly_df
```

```python
vaccinitaly_df["all_urls"] = vaccinitaly_df["urls"] + vaccinitaly_df["quoted_urls"]
```

```python
vaccinitaly_df.drop(urls_to_fix, axis=1, inplace=True)
```

```python
# Turn the urls column in a scores column
```

```python
#url = 'milanotoday.it' # 'imolaoggi.it'
#newsguard_df.query(f"identifier=='{url}' | identifierAlt=='{url}'")["score"].values[0]
```

```python
# Takes a string of URLs like "http://www.donald.com;www.tree.ch;https://nemo.it"
# preprocess extracting top level domain name and return a score for eahc
# missing URLs are handled
def urls_scorer(urls):
    # Urls is a string with items divided by ;
    l = []
    for x in urls.split(";"):
        try:
            l.append(x.split("/")[2].replace("www.", "").lower())
        except IndexError:
            pass
    
    scores = [-1]
    for url in l:
        try:
            score = newsguard_df.query(f"identifier=='{url}' | identifierAlt=='{url}'")["score"].values[0]
            scores.append(score)
        except IndexError: # Not in newsguard df
            continue

    return max(scores) # Here max of the scores is used (replace with any aggregation function like avg or other).
```

```python
# WARNING: Very Slow. Bottleneck. (NOTE: try a parallel implementation with dask?)

tic = time.perf_counter()
vaccinitaly_df["score"] = vaccinitaly_df["all_urls"].apply(urls_scorer)
toc = time.perf_counter()

print(f"Processed: {len(vaccinitaly_df.index)} items in: {round(toc-tic, 4)} seconds.")
```
