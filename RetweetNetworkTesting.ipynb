```python
import pandas as pd

df = covid_df

def original_author_resolver(data_df, content_ID_col, author_ID_col, original_content_ID_col):

    # First, filter out rows where retweeted_status_id is "ORIGIN"
    original_tweets = data_df[data_df[original_content_ID_col] == "ORIGIN"].copy()

    # Rename the columns to avoid conflicts during merge
    original_tweets.rename(columns={author_ID_col: "original_user_id"}, inplace=True)

    # Merge the DataFrame with itself to get the original_user_id for retweets
    result_df = pd.merge(data_df, original_tweets[[content_ID_col, "original_user_id"]],
                         left_on=original_content_ID_col,
                         right_on=content_ID_col,
                         how="left")

    return result_df




# Replace NaN values in the new column with the user_id value for original tweets
#result_df['original_user_id'].fillna(result_df['user_id'], inplace=True)

# Drop the redundant columns
#result_df.drop(columns=['tweet_id_y'], inplace=True)
#result_df.rename(columns={'tweet_id_x': 'tweet_id'}, inplace=True)

# Print the result
#result_df

df = original_author_resolver(covid_df, "tweet_id", "user_id", "retweeted_status_id")
df.tail(50)
```
